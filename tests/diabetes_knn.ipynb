```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier


# Data dari tabel
data = {
 'gula_darah': [99, 110, 153, 128, 88, 156, 99, 109, 120, 140],
 'tekanan_darah': [72, 92, 88, 84, 68, 92, 70, 60, 80, 90],
 'diabetes': [0, 0, 1, 1, 0, 1, 0, 0, 1, 1]
}


# Menambahkan data baru (baris terakhir dengan gula darah 105, tekanan darah 85)
new_data = {
 'gula_darah': 105,
 'tekanan_darah': 85
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Fitur dan label
X = df[['gula_darah', 'tekanan_darah']]
y = df['diabetes']


# Inisialisasi model KNN dengan k=3 (K-Nearest Neighbor)
knn = KNeighborsClassifier(n_neighbors=3)


# Melatih model
knn.fit(X, y)


# Menghitung prediksi untuk data baru
predicted_diabetes = knn.predict([[new_data['gula_darah'], new_data['tekanan_darah']]])
new_data['diabetes'] = predicted_diabetes[0]


data['gula_darah'].append(new_data['gula_darah'])
data['tekanan_darah'].append(new_data['tekanan_darah'])
data['diabetes'].append(new_data['diabetes'])


# Menampilkan hasil
print("Status data baru, Diabetes ?", new_data['diabetes'])


df
```
